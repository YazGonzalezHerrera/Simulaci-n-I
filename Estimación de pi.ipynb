Estimación de π por simulación de Montecarlo

```python
from random import random

N = 1000000
Nc = 0
for i in range(N):
    x = random()
    y = random()
    if x**2 + y**2 < 1:
      Nc = Nc + 1

print(4*Nc/N)
```

```python
from random import random
import matplotlib.pyplot as plt

N = 1000000
Nc = 0
lx = []
ly = []

for i in range(N):
    x = random()
    y = random()
    if x**2 + y**2 < 1:
      Nc = Nc + 1

    if i > 0 and i % 50000 == 0:
        lx.append(i+1)
        ly.append(4*Nc/(i+1))

plt.plot(lx, ly)
plt.show()
```

```python
from random import random
from math import pi
import matplotlib.pyplot as plt

N = 1000000
Nc = 0
lx = []
ly = []
lc = []

for i in range(N):
    x = random()
    y = random()
    if x**2 + y**2 < 1:
      Nc = Nc + 1

    if i > 0 and i % 5000 == 0:
        lx.append(i+1)
        ly.append(4*Nc/(i+1))
        lc.append(pi)

plt.plot(lx, ly)
plt.plot(lx, lc)
plt.show()
```
