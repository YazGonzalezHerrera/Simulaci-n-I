Estimación de $\sqrt{2}$ por simulación de Montecarlo

```python
from random import random
import matplotlib.pyplot as plt

N = 1000000
n = 0
lx = []
ly = []

for i in range(N):
    x = 1 + random()
    if x**2 < 2:
      n = n + 1

    if i > 0 and i % 1000 == 0:
        lx.append(i+1)
        ly.append(1+n/(i+1))

print(1+n/N)
plt.figure(figsize=(4,2))
plt.plot(lx, ly)
plt.show()
```
